````python
"""
clean1 is standard comma seperated with double quote character wrapping each field of records.
Each lines of records in clean1 and first line is the title row  
There are 10 columns of data, other than column2 2 and 3, the rest of columns were number string.
( columns 2 & 3 were the name and type field, should not be converted to numbers )

We are going to convert into variable 'cleanraw' as a list of list object (also array of array).
All number fields from clean1 will be converted to  numbers (float) inside the 'cleanraw'
"""

"""
To read a csv text file from local file system and load it into variable raw_pd.
If we open the csv file directly with a application like Word, we might not see some characters.
raw_pd is the complete collection, including the invisible characters.
Due to being raw, it might have incorrect information or incomplete information.
We will fix it at the later stage
"""

banner("Section 8.1b")

raw_pd = ''
pokedex_file = 'pokedex_basic.csv'
try:
    with open(pokedex_file, 'r') as f:
        raw_pd = f.read()
    print("the file {} has been loaded into memory variable raw_pd".format(pokedex_file))
except:
    print("the file {} might not existed in the local storage or being locked.".format(pokedex_file))
    alert_user
finally:
    f.close()

cleanraw = []

for rows in clean1:
    # Remove quotes
    rows_csv = rows.replace('"', "")
    
    # in case of no values betweem comma such ",," we will make it ",-1,"
    rows_csv = rows_csv.replace(",,", ",-1,")

    # remove single quotes on both sides of the numeric fields, turning '88' into 88
    rows_csv = rows_csv.replace("\''","")

    # Split rows_csv into array of strings
    list_pd = rows_csv.split(",") 
    # List to hold fixed values
    fixed_row = [] 

    # Loop throgh rows to make ammends to cells
    for (i,cell) in enumerate(list_pd):    
        # avoid the name and the type field
        if i==0 or i>2:
            if cell.isdigit():
                # Convert numeric values to floats
                cell = float(cell)  
        fixed_row.append(cell)            
    cleanraw.append(fixed_row)

pprint(cleanraw)
````

````python
"""
To generate a filtered pokedex according to a given filter condition.

Need to prepared:
[A]
Create a dictionary Dict_FieldId to lookup for column number over the column name
Key of Dict_FieldId  Value of Dict_FieldId
(for filter_key)     (for filter_type)
PokedexNumber        0    
Name                 1       
Type                 2       
Total                3        
HP                   4     
Attack               5         
Defense              6          
SpecialAttack        7    
SpecialDefense       8     
Speed                9


For the every field name found inside the filter_options, such as:
filter_options = {
    'PokedexNumber':   6,
}
We can quick slice out the columns since we knows which column number to use.

[B]
Create a list 'ListofRecords' keeping all the record numbers(dictionary key of PokeDex)
This is a complete listing before any filtering runs, the list will gets smaller

What we do:
[A] For each key-value pair in the variable 'filter',
    store the value into 'filter_type' and the key into 'filter_key'
[B] Calculate the column number from Dict_FieldId dictionary using filter_key as a key
[C] Create a new list by filtering according to the criteria.
    Update the list  'ListofRecords' to follow the results.
[D] Repeat step [A] for the next filtering condition

"""

banner("Section 10a")

# loading the cleanraw into PokeDex 
def filtered_pokedex(PokeDex: dict, filter: dict) -> dict:
    
    # Create a dictionary to lookup for column # for respective field name
    # Example, Dict_FieldId[''Defense] returns numeric 6
    Dict_FieldId={x: i for (i, x) in enumerate(cleanraw[0])}

    # List of records number representing the dictionary key of PokeDex
    ListofRecords = [ i+1 for (i, pkd) in enumerate(cleanraw[1:]) ]
    
    # for each filtering condition defined inside variable 'filter'
    for filter_key in filter:
        if filter_key == 'Name':
            filter_type = filter[filter_key]
            pos = Dict_FieldId[filter_key]
            newlist = [r for r in ListofRecords if filter_type in PokeDex[r][pos] ]
            ListofRecords = newlist[:]
            continue
        if filter_key in ['Type','PokedexNumber']:
            filter_type = filter[filter_key]
            pos = Dict_FieldId[filter_key]
            newlist = [r for r in ListofRecords if PokeDex[r][pos] == filter_type]
            ListofRecords = newlist[:]
            continue
        pos = Dict_FieldId[filter_key]
        level = filter[filter_key]
        newlist = [r for r in ListofRecords if PokeDex[r][pos] >= level]
        ListofRecords = newlist[:]
    return[ PokeDex[r] for r in ListofRecords]


# filter based on parameters passed into filtered_pokedex function
filter_options = {
    'Attack':   85,
    'Defense':  60,
    'Name':   'le',
    'Type':     'Electric'
}

PokeDex = PokeDexGen(cleanraw)
pokedex_found = filtered_pokedex(PokeDex,filter=filter_options)
pprint(pokedex_found)
````

### 9.1

What is the population mean and standard deviation of the "Total" attribute for all characters in the Pokedex?




**Mean**

Mean is the average in statistics to describe central tendency of a sample. 
Note that **** is the symbol for mean.

Below formula is mathematical definition:
(where n = the number of records, x are the sample values.)


$$\mu = \frac{1}{n}\sum_{i=1}^{n} {X_i} $$


**Variance**

Variance is to describe the spread of a sample. 
Note that **** is the symbol for variance.

Below formula is mathematical definition:


$$\sigma ^2 = \frac{1}{n} \sum_{i=1}^{n}\left( {X_i} - \mu \right)^2$$



The (xi  ) is called the deviation from the mean.

With this, **Variance** is the squared deviation multiplied by 1 over the number of samples. 

This is why the square root of the variance, , is called the standard deviation.

Furthermore, we can calculate Z-Score for working out probability.
(using CDF cumulative distribution function). 

Below formula is mathematical definition:

$${Z_i} =  \frac{\left( {X_i} - \mu \right)}{\sigma}$$

````python
"""
Extract out the column 3 for the cleanraw ( the 'Total' field ) as a list 'total_attributes'
Use this as the population, to calculate the statistical properties.

We use numpy to calculate the mean and standard deviation

The 'cleanraw' is the population in this case. Note that first row ,#0, is title row.
We takes the 'Total' attributes field is the focus in this case.

Added Get_Total function to make sure data integrity is maintain.
Calculation to check whether 
    Total = HP + Attack + Defense + SpecialAttack + SpecialDefense + Speed
If decrepany found, generate a new list by aggregating the columns 4 to 9.
"""

banner("Section 9.1")

def Get_Total(cleanraw: list) -> list :
    if sum([ -col[3]+col[4]+col[5]+col[6]+col[7]+col[8]+col[9] for col in cleanraw[1::] ])==0:
        return [ col[3] for col in cleanraw[1::] ]
    else:
        return [ col[4]+col[5]+col[6]+col[7]+col[8]+col[9] for col in cleanraw[1::] ]


# Takes the 'Total' attributes from the column #3 from cleanraw
# Row 0 of 'cleanraw' data contains a header, just extract data from row 1 (second row)
# To ensure data integrity, we use Get_Total to check and pull the data from cleanraw
total_attributes = Get_Total(cleanraw)

# Using numpy
population_mean = np.mean(total_attributes)
population_std = np.std(total_attributes)

print("Population mean is {}\nPopulation_std is {}".format(population_mean,population_std) )
````
